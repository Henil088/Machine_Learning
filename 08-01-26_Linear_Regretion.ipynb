```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("H and W (ML) Data.csv")
df
```

```python

df = df.drop(df.index[47])  # drops the row at positional index 47 (48th row)
df
```

```python
df = df[df["Height (in cm)"] > 100]  # remove invalid heigh
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df has columns "Height (in cm)" and "Weight (in kg)"

sns.set(style="whitegrid")  # optional, makes it look nicer

# Scatter plot
sns.scatterplot(data=df, x="Height (in cm)", y="Weight (in kg)", color="blue", s=100)

# Labels and title
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight")

plt.show()
```

```python
x = df["Height (in cm)"].values
y = df["Weight (in kg)"].values

x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

m = numerator / denominator
c = y_mean - m * x_mean

m, c
```

```python
y_pred = m * x + c
```

```python
sns.set(style="whitegrid")
sns.scatterplot(data=df, x="Height", y="Weight", label="Actual Data")
sns.lineplot(data=df, x="Height", y="Weight_pred", color="red", label="Regression Line")

plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend()
plt.show()

```

```python
from sklearn.linear_model import LinearRegression

X = np.array([170, 160, 165]).reshape(-1, 1)
Y = np.array([160, 165, 170])

model = LinearRegression()
model.fit(X, Y)

y_predicted = model.predict(X)


df_final = pd.DataFrame({
    "X": X.flatten(),
    "Y_predicted": y_predicted,
    "Error": Y - y_predicted
})


df_final["Y"] = Y

print(df_final)

```
