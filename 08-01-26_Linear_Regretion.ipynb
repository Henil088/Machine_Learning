```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("H and W (ML) Data.csv")
df
```

```python
df = df[df["Height (in cm)"] > 100]  # remove invalid heigh
```

```python
plt.scatter(df["Height (in cm)"], df["Weight (in kg)"])
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight")
plt.show()
```

```python
x = df["Height (in cm)"].values
y = df["Weight (in kg)"].values

x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

m = numerator / denominator
c = y_mean - m * x_mean

m, c
```

```python
y_pred = m * x + c
```

```python
plt.scatter(x, y, label="Actual Data")
plt.plot(x, y_pred, color="red", label="Regression Line")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend()
plt.show()
```

```python
print(f"Weight = {m:.2f} * Height + {c:.2f}")
```

```python
def predict_weight(height):
    return m * height + c

predict_weight(170)
```
