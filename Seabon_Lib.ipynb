```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# sample dataset 

df = pd.DataFrame(
    {
        "x": [1,2,3,4,5,6],
        "y": [5,7,6,8,7,9],
        "category": ["A", "A", "B", "B", "C", "C"]
    }
)
print(df)
```

```python
sns.scatterplot(x = "x", y = "y", data=df)
plt.show()
```

```python
sns.lineplot(x = "x", y = "y", data=df)
plt.show()
```

```python
sns.regplot(x = "x", y = "y", data=df)
plt.show()
```

```python
sns.lmplot(x = "x", y = "y", data=df)
plt.show()
```

```python
sns.barplot(x="category", y="y", data=df)
plt.show()
```

```python
sns.boxplot(x="category", y="y", data=df)
plt.show()
```

```python
sns.pointplot(x="category", y="y", data=df)
plt.show()
```

```python
sns.catplot(x="category", y="y", data=df)
plt.show()
```

```python
sns.heatmap(df[["x", "y"]].corr(), annot= True)
plt.show()
```
